```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.columns
```

```python
df.info()
```

```python
list = ['fueltype', 'aspiration',
       'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'enginetype',
       'cylindernumber', 'fuelsystem']
for col in list:
    enc = LabelEncoder()
    df[col] = df[col].astype('category')
    encoded = enc.fit_transform(df[[col]])
    df[col] = encoded
```

```python
df.info()
df['price'] = df['price'].astype('int')
```

```python
[df['fuelsystem'].unique() for col in df.columns]
```

```python
x = df.drop(['car_ID','CarName','symboling','price','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke'],axis=1)
pd.set_option('display.max_columns', None)
x.head()
x.columns
x[x['enginelocation']!=0]
x
#fueltype gas=1,diesel=0
#aspiration standard=1,turbo=0
#doornumber two=1, four=0
#carbody convertible=0,hardtop=1,hatchback=2,sedan=3,wagon=4
#drivewheel rearwheeldrive=2,front wheel drive=1, 4 wheel drive=0
#enginelocation front=0,rear=1
#enginetype dual overhead camshaft(dohc)=0, overhead camshaft valve(ohcv)=5, overhead camshaft(ohc)=3, l=2, rotor=6, ohcf=4, dohcv=1
#cylindernumber 4cylinders=2,6cylinder=3,5cylinder=1,3c=4,12c=5,2c=6,8c=0
#fuelsystem mpfi=5,2bbl=1,mfi=4,1bbl=0,spfi=7,4bbl=2,id=3,spdi=6

print(x['highwaympg'].min())
print(x['highwaympg'].max())
print(x)
```
